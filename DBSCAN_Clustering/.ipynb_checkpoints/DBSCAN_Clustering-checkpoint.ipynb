## We use the moon dataset to study DBSCAN clustring algorithm

```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.model_selection import GridSearchCV
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X, y = make_moons(n_samples = 1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

Some instances have a cluster index equal to -1: this means that they are considered as anomalies by the algorithm.   

The indices of the core instances are available in the core_sample_indices_ instance variable.  
The core instances themselves are available in the components_ instance variable

```python
len(dbscan.core_sample_indices_)
```

```python
dbscan.core_sample_indices_
```

```python
len(dbscan.components_)
```

```python
dbscan.components_
```

```python
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]
    
    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=20, c=dbscan.labels_[core_mask])
    plt.scatter(anomalies[:, 0], anomalies[:, 1],
                c="r", marker="x", s=100)
    plt.scatter(non_cores[:, 0], non_cores[:, 1], c=dbscan.labels_[non_core_mask], marker=".")
    if show_xlabels:
        plt.xlabel("$x_1$", fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel("$x_2$", fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title("eps={:.2f}, min_samples={}".format(dbscan.eps, dbscan.min_samples), fontsize=14)
```

```python
dbscan2 = DBSCAN(eps=0.2, min_samples=5)
dbscan2.fit(X)
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(121)
plot_dbscan(dbscan, X, size=100)

plt.subplot(122)
plot_dbscan(dbscan2, X, size=600, show_ylabels=False)

plt.show()
```

In the First image, we apply the algorithm DBSCAN for eps=0.05 and generates 7 clusters  
By modifying eps to 0.2, we obtain a better result clear 2 clusters in the second image.

DBSCAN class does not have a predict() method, although it has a fit_predict() method. In other words, it cannot predict which
cluster a new instance belongs to.  

There fore, we can train a KNN classifier using DBSCAN results core instance (components as features X and labels as target y)

```python
knn = KNeighborsClassifier(n_neighbors = 50)
knn.fit(dbscan2.components_, dbscan2.labels_[dbscan2.core_sample_indices_])
```

```python
X_new =np.array([[-0.5, 0], [0, 0.5], [1, -0.1], [2, 1]])
```

```python
knn.predict(X_new)
```

```python
knn.predict_proba(X_new)
```

There is no anomaly in the KNN’s training set, thus the classifier always chooses a cluster, even when that cluster is far away. However, we can introduce a maximum distance, in which case the two
instances that are far away from both clusters are classified as anomalies.

If we use the kneighbors() method of the KNeighborsClassifier: given a set of instances, it returns the distances and the indices of the k nearest neighbors in the
training set.

If the density varies significantly across the clusters, it can be impossible for the DBSCAN algorithm to capture all the clusters properly
